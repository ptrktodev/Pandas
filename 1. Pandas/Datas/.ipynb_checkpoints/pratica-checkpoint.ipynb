```python
import seaborn as sns
```

```python
df_titanic = sns.load_dataset('titanic')
df_titanic
```

```python
df_titanic.set_index('class').loc['Third'].query("`sex` == 'female'")['age'].mean()
```

```python
paleta = ['#5b75f3', '#f706b6']
sns.boxplot(data=df_titanic, x='class', y='age', hue='sex', palette=paleta )
```
